```python
import pandas as pd
movies = pd.read_csv('movies.csv')
display(movies.info())
```

```python
ratings1 = pd.read_csv('ratings1.csv')
display(ratings1.nunique())
```

```python
dates = pd.read_csv('dates.csv')
dates.value_counts()
```

```python
dates.sort_values(by='date').head(10)
```

```python
dates['date'] = pd.to_datetime(dates['date'])
dates['Year'] = dates['date'].dt.year
display(dates['Year'].mode())
```

```python
ratings2 = pd.read_csv('ratings2.csv')

ratings = pd.concat([ratings1, ratings2], ignore_index = True)
display(ratings)
```

```python
print('Число строк в таблице ratings: ', ratings.shape[0])
print('Число строк в таблице dates: ', dates.shape[0])
print(ratings.shape[0] == dates.shape[0])
```

```python
ratings = ratings.drop_duplicates(ignore_index=True)
print('Число строк в таблице ratings: ', ratings.shape[0])
```

```python
ratings_dates = pd.concat([ratings, dates], axis=1)
display(ratings_dates.tail(7))
```

```python
import os
```

```python
joined_false = ratings_dates.join(
    movies,
    rsuffix='_right',
    how='left'
)
display(joined_false)
```

```python
joined = ratings_dates.join(
    movies.set_index('movieId'),
    on='movieId',
    how='left'
)
display(joined.head())
display(joined.tail())
```

```python
merged = ratings_dates.merge(
    movies,
    on='movieId',
    how='left'
)
display(merged.head())
```

```python
print('Число строк в таблице ratings_dates: ', ratings_dates.shape[0])
print('Число строк в таблице merged: ', merged.shape[0])
print(ratings_dates.shape[0] == merged.shape[0])
```

```python
print('Число колонок в таблице ratings_dates: ', ratings_dates.shape[1])
print('Число колонок в таблице merged: ', merged.shape[1])
print(ratings_dates.shape[0] == merged.shape[1])
```
