```python
import pandas as pd
ratings_movies = pd.read_csv('ratings_movies.csv')
display(ratings_movies.info())
```

```python
display(ratings_movies.head())
```

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
display(ratings_movies.info())
```

```python
mask1 = ratings_movies['year_release'] == 1999
ratings_movies[mask1].groupby('title')['rating'].mean().sort_values()
```

```python
mask2 = ratings_movies['year_release'] == 2010
ratings_movies[mask2].groupby('genres')['rating'].mean().sort_values()
```

```python
mask3 = ratings_movies['year_release'] == 2018
ratings_movies[mask3].groupby('genres')['rating'].agg(['count', 'mean']
).sort_values( ['count', 'mean'], ascending = [True, False])
```

```python
ratings_movies.groupby('userId')['rating'].agg(
    ['count', 'mean']
).sort_values( ['count', 'mean'], ascending = [True, False])
```

```python
ratings_movies.groupby('genres')['rating'].mean().sort_values()
```

```python
ratings_movies['year_rating'] = ratings_movies['rating'].mean()
display(ratings_movies.head())
```
